```python
tickers_January = pd.read_csv('data/all_tickers_yfinance_January2021.csv', index_col = 0)

tickers_January.head()
tickers = tickers_January.index
```

###  Información de Yahoo Finance de cada ticker

Esto tarda aproximadamente 5 horas

```python
%%time

# Initial ticker
first_ticker = yf.Ticker(tickers[0])

ticker_yf_data = pd.Series(first_ticker.info, name = tickers[0]).to_frame().T

for i, tck in enumerate(tickers):
    if float(i) % 100 == 0:
        print('Faltan {} tickers...'.format(len(tickers) - i))
    try:
        ticker_yf_data.loc["{}".format(tck)] = pd.Series(yf.Ticker(tck).info)
    except:
        continue
```

#### Se filtran solo equities

```python
ticker_yf_data = ticker_yf_data[ticker_yf_data['quoteType'] == 'EQUITY']
ticker_yf_data.head()
```

#### Guarda archivo

Se genera nombre...

```python
today = pd.to_datetime("today")

filename ='data/all_tickers_yfinance_' + today.month_name() + str(today.year) + '.csv'

ticker_yf_data.to_csv(filename)
```
